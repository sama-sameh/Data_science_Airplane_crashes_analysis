# Data cleaning

```python
df.drop(['Flight #','Registration','cn/In','Time'], axis = 1,inplace = True)
```

```python
df.columns = df.columns.str.lower()
```

```python
df.head()
```

## Handling Missing Values
* If col has missing < 5 % : Drop Rows with missing values
* If col has missing > 40 % : Drop the column
* If col has missing (5-40 %): Filling

```python
df.isna().mean()[df.isna().mean()>0]*100
```

```python
df['location'] = df['location'].fillna(df['location'].mode()[0])
df['operator'] = df['operator'].fillna(df['operator'].mode()[0])
df['route'] = df['route'].fillna(df['route'].mode()[0])
df['type'] = df['type'].fillna(df['type'].mode()[0])
```

```python
# if there are many aboards of same type
df['aboard'] = df['aboard'].fillna(df.groupby('type')['aboard'].transform('median'))

# Ensure Aboard is at least Fatalities
df['aboard'] = df.apply(lambda row: max(row['aboard'], row['fatalities']) if pd.notna(row['fatalities']) else row['aboard'], axis=1)
```

```python
# if there are one aboards of a type
df['aboard'] = df['aboard'].fillna(df['aboard'].median())
df['aboard'] = df.apply(lambda row: max(row['aboard'], row['fatalities']) if pd.notna(row['fatalities']) else row['aboard'], axis=1)
```

```python
df.dropna(subset = 'ground',inplace = True)
```

```python
df['summary'] = df.groupby(['type', 'operator'])['summary'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else x)
df['summary'] = df['summary'].fillna(df['summary'].mode()[0])
```

```python
df.aboard = df.aboard.astype('int')
df.fatalities = df.fatalities.astype('int')
df.aboard = df.aboard.astype('int')
df.ground = df.ground.astype('int')
```

## Feature extraction 

```python
def classify_crash(row):
    if row.aboard == row.fatalities:
        return "All Died"
    return "Some Survived"

df['status'] = df.apply(classify_crash, axis = 1)
```

```python
def get_year(row):
    return row.date.split('/')[2]
df['year'] = df.apply(get_year, axis = 1)
```
